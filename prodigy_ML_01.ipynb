data preprocessing for the records in train dataframe

```python

tr_LivArea_Median = math.floor(df_train['GrLivArea'].median())
df_train['GrLivArea'].fillna(tr_LivArea_Median, inplace=True)

tr_BsmtSF_Median = math.floor(df_train['TotalBsmtSF'].median())
df_train['TotalBsmtSF'].fillna(tr_BsmtSF_Median, inplace=True)

tr_GarageArea_Median = math.floor(df_train['GarageArea'].median())
df_train['GarageArea'].fillna(tr_GarageArea_Median, inplace=True)

tr_WoodDeck_Median = math.floor(df_train['WoodDeckSF'].median())
df_train['WoodDeckSF'].fillna(tr_WoodDeck_Median, inplace=True)

tr_OpenPorch_Median = math.floor(df_train['OpenPorchSF'].median())
df_train['OpenPorchSF'].fillna(tr_OpenPorch_Median, inplace=True)

tr_EnclosedPorch_Median = math.floor(df_train['EnclosedPorch'].median())
df_train['EnclosedPorch'].fillna(tr_EnclosedPorch_Median, inplace=True)

tr_3SsnPorch_Median = math.floor(df_train['3SsnPorch'].median())
df_train['3SsnPorch'].fillna(tr_3SsnPorch_Median, inplace=True)

tr_ScreenPorch_Median = math.floor(df_train['ScreenPorch'].median())
df_train['ScreenPorch'].fillna(tr_ScreenPorch_Median, inplace=True)

tr_PoolArea_Median = math.floor(df_train['PoolArea'].median())
df_train['PoolArea'].fillna(tr_PoolArea_Median, inplace=True)

tr_Bsmt_FB_Median = math.floor(df_train['BsmtFullBath'].median())
df_train['BsmtFullBath'].fillna(tr_Bsmt_FB_Median, inplace=True)

tr_Bsmt_HB_Median = math.floor(df_train['BsmtHalfBath'].median())
df_train['BsmtHalfBath'].fillna(tr_Bsmt_HB_Median, inplace=True)

tr_FB_Median = math.floor(df_train['FullBath'].median())
df_train['FullBath'].fillna(tr_FB_Median, inplace=True)

tr_HB_Median = math.floor(df_train['HalfBath'].median())
df_train['HalfBath'].fillna(tr_HB_Median, inplace=True)

tr_Bedroom_Median = math.floor(df_train['BedroomAbvGr'].median())
df_train['BedroomAbvGr'].fillna(tr_Bedroom_Median, inplace=True)

df_train
```

```python
# Edit fields in train dataframe
df_train['TotSqFt'] = df_train['GrLivArea'] + df_train['TotalBsmtSF'] + df_train['GarageArea'] + df_train['WoodDeckSF'] + df_train['OpenPorchSF'] + df_train['EnclosedPorch'] + df_train['3SsnPorch'] + df_train['ScreenPorch'] + df_train['PoolArea']
df_train['TotBaths'] = df_train['BsmtFullBath'] + df_train['BsmtHalfBath'] + df_train['FullBath'] + df_train['HalfBath']
df_train = df_train.drop(columns=df_train.columns.difference(['Id', 'TotSqFt', 'TotBaths', 'BedroomAbvGr', 'SalePrice']))
df_train
```

data preprocessing for the records in test dataframe

```python
te_Liv_Area_Median = math.floor(df_test['GrLivArea'].median())
df_test['GrLivArea'].fillna(te_Liv_Area_Median, inplace=True)

te_BsmtSF_Median = math.floor(df_test['TotalBsmtSF'].median())
df_test['TotalBsmtSF'].fillna(tr_BsmtSF_Median, inplace=True)

te_GarageArea_Median = math.floor(df_test['GarageArea'].median())
df_test['GarageArea'].fillna(tr_GarageArea_Median, inplace=True)

te_WoodDeck_Median = math.floor(df_test['WoodDeckSF'].median())
df_test['WoodDeckSF'].fillna(tr_WoodDeck_Median, inplace=True)

te_OpenPorch_Median = math.floor(df_test['OpenPorchSF'].median())
df_test['OpenPorchSF'].fillna(tr_OpenPorch_Median, inplace=True)

te_EnclosedPorch_Median = math.floor(df_test['EnclosedPorch'].median())
df_test['EnclosedPorch'].fillna(tr_EnclosedPorch_Median, inplace=True)

te_3SsnPorch_Median = math.floor(df_test['3SsnPorch'].median())
df_test['3SsnPorch'].fillna(tr_3SsnPorch_Median, inplace=True)

te_ScreenPorch_Median = math.floor(df_test['ScreenPorch'].median())
df_test['ScreenPorch'].fillna(tr_ScreenPorch_Median, inplace=True)

te_PoolArea_Median = math.floor(df_test['PoolArea'].median())
df_test['PoolArea'].fillna(tr_PoolArea_Median, inplace=True)

te_Bsmt_FB_Median = math.floor(df_test['BsmtFullBath'].median())
df_test['BsmtFullBath'].fillna(te_Bsmt_FB_Median, inplace=True)

te_Bsmt_HB_Median = math.floor(df_test['BsmtHalfBath'].median())
df_test['BsmtHalfBath'].fillna(te_Bsmt_HB_Median, inplace=True)

te_FB_Median = math.floor(df_test['FullBath'].median())
df_test['FullBath'].fillna(te_FB_Median, inplace=True)

te_HB_Median = math.floor(df_test['HalfBath'].median())
df_test['HalfBath'].fillna(te_HB_Median, inplace=True)

te_Bedroom_Median = math.floor(df_test['BedroomAbvGr'].median())
df_test['BedroomAbvGr'].fillna(te_Bedroom_Median, inplace=True)

df_test
```

```python
# Edit fields in test dataframe
df_test['TotSqFt'] = df_test['GrLivArea'] + df_test['TotalBsmtSF'] + df_test['GarageArea'] + df_test['WoodDeckSF'] + df_test['OpenPorchSF'] + df_test['EnclosedPorch'] + df_test['3SsnPorch'] + df_test['ScreenPorch'] + df_test['PoolArea']
df_test['TotBaths'] = df_test['BsmtFullBath'] + df_test['BsmtHalfBath'] + df_test['FullBath'] + df_test['HalfBath']
df_test = df_test.drop(columns=df_test.columns.difference(['Id', 'TotSqFt', 'TotBaths', 'BedroomAbvGr', 'SalePrice']))
df_test
```

Specify features and target variables from training dataframe

```python
x = df_train[['TotSqFt', 'TotBaths', 'BedroomAbvGr']]
y = df_train['SalePrice']
```

```python
# Split the data into training and testing sets
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=42)
```

```python
# Train the linear regression model
lr_model = LinearRegression()
lr_model.fit(x_train, y_train)
```

```python
# Let's see the coefficient and intercept values of the equation y = mx + c
print(f"\nCoefficients : {lr_model.coef_}\nIntercept : {lr_model.intercept_}")
```

```python
# Let's make predictions on the test set of the split
y_pred = lr_model.predict(x_test)
```

```python
# Let's evaluate the model
mse = mean_squared_error(y_test, y_pred)
print(f"\nMean Squared Error on Test Set: {mse}")

rmse = math.sqrt(mse)
print(f"\nDeviation : {rmse}")
```

```python
# Let's make predictions on the test dataframe
test_predictions = lr_model.predict(df_test[['TotSqFt', 'TotBaths', 'BedroomAbvGr']])
```
